```python
arr = np.array([
[2, 3, 5],
[14, 6, 8],
[11, 13, 17],
[7, 10, 13]
])
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n) + 1)):
        if n % i == 0:
            return False
    return True
```

```python
def rows_at_least_one_prime_number(arr):
    prime_mask = np.vectorize(is_prime)(arr)
    row_mask = np.any(prime_mask, axis = 1)
    return arr[row_mask]
```

```python
print(rows_at_least_one_prime_number(arr))
```

```python
def checkerboard():
    board = np.zeros((8,8),dtype=int)
    board[::2, ::2] = 1
    board[1::2, 1::2] = 1
    return board

print(checkerboard())
```

```python
def reverse_checkerboard():
    board = np.zeros((8,8),dtype=int)
    board[::2, 1::2] = 1
    board[1::2, ::2] = 1
    return board

print(reverse_checkerboard())
```

```python
def expansion(arr, spaces):
    space_number = " " * spaces
    return np.array([space_number.join(word) for word in arr])
```

```python
universe = np.array(["galaxy", "clusters"])
```

```python
expansion(universe, 1)
```

```python
expansion(universe, 2)
```

```python
def secondDimmest(stars):
    sorted_row = np.sort(stars, axis=1)
    return sorted_row[:,-2]
```

```python
np.random.seed(123)
stars = np.random.randint(500, 2000, (5, 5))
print(stars)
```

```python
secondDimmest(stars)
```
