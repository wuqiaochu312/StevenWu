```python
list = [20, 27, 9, 12, 31, 17, 4, 50, 39, 35]
```

```python
arr = np.array(list)
```

```python
print("arr * 4:", arr * 4)
```

```python
print("mean:", np.mean(arr))
```

```python
print("median:", np.median(arr))
```

```python
print("standard deviation:", np.std(arr))
```

```python
min_index = np.argmin(arr)
max_index = np.argmax(arr)

print("index of smallest value:", min_index)
print("index of largest value:", max_index)
```

```python
print("smallest value:", arr[min_index])
print("largest value:", arr[max_index])
```

```python
print("smallest value:", np.min(arr))
print("largest value:", np.max(arr))
```

```python
greater_than_20 = arr[arr > 20]
```

```python
print("new_arr:", greater_than_20)
```

```python
matrix = np.array([
[5, 12, 17, 9],
[8, 21, 14, 3],
[6, 13, 19, 11]
])
```

```python
print("shape:", matrix.shape)
print("row:", matrix.shape[0])
print("column:", matrix.shape[1])
```

```python
print("third column:", matrix[:,2])
```

```python
print("second row:", matrix[1])
```

```python
print("matrix * 2:", matrix * 2)
```

```python
print("mean of each column:", np.mean(matrix,axis=0))
```

```python
print("overall mean:", np.mean(matrix))
```

```python
print("max in each row:", np.max(matrix, axis=1))
print("min in each column:", np.min(matrix, axis=0))
```
